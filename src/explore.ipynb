```python
import sqlite3

con = sqlite3.connect('classDatabase.db') #to create a file

con.execute("""
            
    create table books (
        book_id int unique primary key not null, 
        title text,
        author text,
        pages int,
        rating float    
    )
    

""") 

```

```python
import sqlite3

con = sqlite3.connect('classDatabase.db') #to create a file

con.execute("""
            
    insert into books (book_id, title, author, pages, rating) values(1, "bloody meridian", "C. Mcc.", 200, 9.5)
    
    

""") 
```

```python
cur = con.execute("""
                  
    select book_id, title from books
                  
""")

for result in cur:
    print(result)
```
